```python
import itertools
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, accuracy_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train = pd.read_csv('datasets/input/datatrain.csv')
test = pd.read_csv('datasets/input/datatest.csv')
```

```python
def plot_confusion_matrix(cm, classes):
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = 'd' 
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
```

```python
selected_x1 = train[['mtot','fermi','banden','nelectrons']]
plt.title('Selected feature correlation')
corr_df = pd.concat([selected_x1, train['class']], axis=1)
sns.heatmap(corr_df.corr(), annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm')
```

```python
x1 = train.drop(columns=['class','structure'])
y1 = train['class'].values
xtest1 = test.drop(columns=['class','structure'])
ytest1 = test['class'].values
i = 3
while (i <= 5):
    for j in range(1,3):
        if (j == 1):
            print('Manhattan Distance: ')
        else:
            print('Euclidean Distance: ')
        knn = KNeighborsClassifier(n_neighbors=i,p=j)
        knn.fit(x1,y1)
        hasil = knn.predict(xtest1)
        print(hasil[:])
        print('K = ',+i)
        cfm = confusion_matrix(ytest1,hasil)
        print(cfm)
        accuracy = accuracy_score(hasil,ytest1)
        print('akurasi: {}'.format(accuracy))
        total = sum(sum(cfm))
        sensitivity = cfm[0,0]/(cfm[0,0]+cfm[0,1])
        print('sensitivity: ',sensitivity)
        specificity = cfm[1,1]/(cfm[1,0]+cfm[1,1])
        print('specificity: ',specificity)
        print('')
    i+=2
```

```python
x1 = train.drop(columns=['class','structure','mtot'])
y1 = train['class'].values
xtest1 = test.drop(columns=['class','structure','mtot'])
ytest1 = test['class'].values
i = 3
while (i <= 9):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x1,y1)
    hasil = knn.predict(xtest1)
    print(hasil[:])
    print('K = ',+i)
    print(confusion_matrix(ytest1,hasil))
    print(classification_report(ytest1,hasil))
    i+=2
```
